```python
X= df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]#X Array is defined
```

```python
y= df['Price']#Y Array is defined
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm= LinearRegression()
```

```python
lm.fit(X_train,y_train)#Training the model
```

```python
#Evaluate the Model
```

```python
lm.intercept_
```

```python
cdf=pd.DataFrame(lm.coef_,X.columns,columns=['Coef'])
```

```python
cdf
```

```python
#Predictions
```

```python
predictions=lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.distplot((y_test-predictions))
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test,predictions)#mean absolute error
```

```python
metrics.mean_squared_error(y_test,predictions)#mean square error
```

```python
np.sqrt(metrics.mean_squared_error(y_test,predictions))#roor mean square model
```

```python
#Training and Testing the data
```

```python
X = df1[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
```

```python
y= df1['Yearly Amount Spent']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#Train the model
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
# The coefficients
print('Coefficients: \n', lm.coef_)
```

```python
cdf1 = pd.DataFrame(lm.coef_,X.columns,columns=["Coef"])
```

```python
cdf1
```

```python
lm.intercept_
```

```python
#Predictions
```

```python
predictions=lm.predict(X_test)
```

```python
y_test
```

```python
sns.scatterplot(y_test,predictions)
```

```python
#Evaluate the model
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
sns.distplot((y_test-predictions))
```
